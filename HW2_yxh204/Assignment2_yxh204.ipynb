```python
data_2 = pd.read_csv("y43c-5n92")
```

```python
data_2.head()
```

```python
data_2 = data_2[["Date", "Average 24hrTurbidity(NTU)"]]
data_2
```

```python
strip_date = lambda x: x[:10]
data_2.Date = data_2.Date.apply(strip_date)
data_2.Date = pd.to_datetime(data_2.Date, format = "%m/%d/%Y")
data_2.columns = ["Date", "Avg24hrTurbidity"]
```

```python
import matplotlib.dates as mdates
fig = plt.figure(figsize = (8,8))
ax = plt.subplot(111)

fig.autofmt_xdate()
plt.plot(data_2.Date, data_2.Avg24hrTurbidity)

xfmt = mdates.DateFormatter('%m/%d')
ax.xaxis.set_major_formatter(xfmt)

plt.ylabel("Average 24hr Turbidity(NTU)"); plt.xlabel("Date")
plt.title("Time Series Plot of Average 24hr Turbidity(NTU)")
plt.show()
```

<center>*Figure 2: Time Series Plot of Average 24hr Turbidity(NTU)*</center>
